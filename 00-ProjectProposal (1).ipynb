## Research Question


How is average weekly physical activity (minutes per week) associated with adult obesity prevalence, diabetes prevalence, and mental health(by a self designed metrics from an online data base) across U.S. states?


## Background and Prior Work


Health outcomes vary across the United States and are influenced by a combination of biological and lifestyle factors. Conditions like diabetes, obesity, and mental health is not attributed to a single cause. Factors like genetics, medical conditions, access to healthcare and the daily routine of an individual all play a role. One thing we can say about is that physical activity is one of the factors that can easily improve such conditions when time is dedicated to it weekly, and our goal is to examine how physical activity is associated with adult obesity prevalence, diabetes prevalence, and mental health across the U.S.


Physical activity plays an important role in maintaining both physical and mental health and it is recognized as an important health factor. Physical activity includes everyday movements such as walking, running, stretching, or even having a structured workout. According to the World Health Organization (WHO), individuals should engage in at least 150 minutes of moderate to intense physical activity per week to support health outcomes <a name="cite_ref-1"></a><a href="#cite_note-1">1</a>. Despite the recommendations from this organization, the amount of time or levels of physical activity varies across the individuals in the U.S. due to the differences they have in their work schedules, access to resources, and personal life.




Previous research has shown the relationship between physical activity and its health outcomes, such as obesity and diabetes. The Center for Disease Control and Prevention (CDC) has reported that obesity and diabetes prevalence has steadily increased across the U.S. over the past decades <a name="cite_ref-2"></a><a href="#cite_note-2">2</a>. We noticed that these trends are influenced by a variety of factors which include diet quality, limited physical activity, metabolic conditions like thyroid conditions, or even the individual's metabolism. However, in this project, we focus on physical activity as the key variable and examine how it is associated with obesity and diabetes.


Moving along with physical health, there is also prior work that examined the relationship between physical activity and mental health outcomes. Research from the Harvard T.H. Chan School of Public Health found that even modest amounts of physical activity, such as 15 minutes of exercise per day or even one hour of walking, is associated with a significant reduction in the risk of major depression <a name="cite_ref-3"></a><a href="#cite_note-3">3</a>. This supports the inclusion of mental health as an outcome variable in our project, along with obesity and diabetes when we examine how physical activity varies across the U.S. 


Our group has noticed that the existing research focuses more on the mental and health benefits for an individual. Our goal is to show how physical activity helps in a broader geographic level  which is in the U.S. The differences in physical activity rates across the states may help explain the observed variation in obesity prevalence, diabetes prevalence and mental health indicators. This project aims to identify patterns that may help inform public health strategies and highlight the disparities in health and mental outcomes related to physical activity.

## Hypothesis



We predict a strong negative correlation between minutes of physical activity and obesity rates and diabetes diagnosis rates in the United States, due to commonly reported sentiments that physical activity improves physical health. We also predict that higher minutes of physical activity have a strong positive correlation with self-reported mood, based on anecdotal evidence we have about positive emotions post-physical activity.

## Data

Ideal dataset (one per health outcome; we plan to synthesize three datasets)
- Core purpose: measure the association between minutes of physical activity and (a) diabetes status, (b) obesity (BMI / body-fat), and (c) self-reported mood / mental-health score.
- Required variables (per observation / participant):
  1. Unique participant ID (non-identifying)
  2. Timestamp / survey date and U.S. state (for aggregation & linkage)
  3. Physical activity: minutes of activity per day or per week (numeric). Ideally split by intensity (light/moderate/vigorous) and, optionally, a free-text/activity-type field.
  4. Height and weight (to compute BMI) and/or body fat percentage
  5. Clinical/diagnostic label or outcome variable:
     - Diabetes dataset: diabetes diagnosis (none / prediabetes / diabetes) and, if available, lab measures (A1C, fasting glucose)
     - Obesity dataset: measured BMI or body-fat percentage and categories (normal/overweight/obese)
     - Mental health dataset: validated self-report score (e.g., PHQ-9, GAD-7, or composite mood score)
  6. Key covariates: age, sex, race/ethnicity, socioeconomic indicator (education/income), smoking status, and comorbidities or medication use
  7. (Optional) Device/source metadata (self-report vs. wearable) and sampling method notes
- Minimum sample size: at least 10,000 independent observations per outcome dataset to ensure statistical power and permit subgroup analyses (e.g., by age, sex, or state).
- How collected: prospective clinical/epidemiological studies or large-scale national surveys carried out in the U.S., or repositories of wearable-device studies where participants consented to research use. Measurement should be standardized (e.g., validated survey items or device-calibrated minutes) and accompanied by documentation describing sampling frame and inclusion/exclusion criteria.
- Storage & organization: CSV / Parquet tables with one row per participant-observation, a data dictionary (README) describing each field, and separate metadata files with provenance and access requirements. Use normalized keys to enable linking state-level covariates or external population measures. Keep raw data read-only and produce a cleaned analysis dataset with versioned scripts (Jupyter / Python notebooks) so every transformation is reproducible.

Real datasets that could be useful (short descriptions and important variables)

1) Fitness/device-based activity linked to BMI and steps — JMIR wearable study  
- Where: Study published in JMIR describing Fitbit-derived daily steps and minutes by intensity for ~13,947 consenting U.S. participants; access to underlying research data may require application/permission through the study’s data access channel or affiliated research program.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)  
- Important variables: daily step counts; minutes per day of sedentary/light/moderate/vigorous activity; participant age, weight, height, BMI; timestamps. This dataset matches our ideal activity measurement (minutes/day and intensity) and contains measured anthropometrics, making it a strong candidate for the obesity/physical-activity analysis. Access may require data-use permission.

2) CDC Behavioral Risk Factor Surveillance System (BRFSS) — state-level physical activity and health indicators  
- Where: The CDC BRFSS provides annual, state-representative survey data for U.S. adults and is publicly available from the CDC; BRFSS includes large sample sizes (hundreds of thousands of respondents across years) and state identifiers for aggregation.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)  
- Important variables: self-reported physical activity (some years include minutes/week or frequency), diagnosed diabetes (yes/no), BMI (derived from self-reported height and weight), demographic variables (age, sex, race, education, income), and state. BRFSS data are ideal for state-level analyses and for building large-sample diabetes/obesity datasets, though activity measures may be self-reported and sometimes categorical rather than precise minutes.

3) NHANES (National Health and Nutrition Examination Survey) — measured anthropometrics, biomarkers, and activity modules  
- Where: NHANES provides measured height/weight, biomarkers (A1C, glucose), and in some cycles accelerometer or questionnaire-based physical-activity modules; data are publicly available via CDC with rich documentation and the possibility to combine multiple cycles to increase sample size.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)  
- Important variables: objectively measured height and weight (BMI), lab measures relevant to diabetes (A1C, fasting glucose), self-reported or device-measured physical activity (minutes by intensity in some modules), and demographics. NHANES’ measured outcomes increase validity, and pooling cycles can approach the sample-size goals for some analyses; however, sample design and weights must be handled properly.

4) Diabetes health indicators (Kaggle mirror of CDC survey-derived dataset) — alexteboul / Diabetes Health Indicators Dataset  
- Where: A Kaggle dataset aggregating CDC survey indicators (mirror on Kaggle). The dataset is openly accessible on Kaggle but may carry the caveat that some fields are preprocessed for Kaggle use; review original documentation if using it for publication-quality work.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)  
- Important variables: diabetes diagnosis (binary or multiclass), BMI or weight/height, smoking and other risk factors. This dataset has very large n (hundreds of thousands when derived from CDC sources) and is immediately useful for diabetes risk-factor analyses, but its physical-activity field is binary (any activity in the past 30 days), which does not match our ideal minutes-based activity variable; it can still provide useful covariate and outcome modeling.

5) Mental health × physical activity (Kaggle dataset by Safiya Fatima)  
- Where: Kaggle user dataset that includes weekly physical activity time and mental-health indicators; data are available on Kaggle and can be downloaded after agreeing to Kaggle’s terms.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)  
- Important variables: reported minutes of exercise per week, occupation, and several mental-health related columns. This dataset directly reports weekly exercise time (close to our minutes-per-week ideal) and mental-health measures; however, the dataset is relatively small and contains missing values in text/activity-type fields, so it will require cleaning. We plan to use such a dataset in combination with larger, complementary sources to improve statistical power and generalizability.

How we will combine and use these resources
- Prioritize datasets that contain minutes-based activity measures (device-derived or validated self-report) for the primary exposure variable. For outcomes where minutes are unavailable (e.g., certain BRFSS variables), use them as supplementary/state-level confirmatory analyses and be explicit about measurement differences.  
- Use NHANES and device-based cohorts where available for analyses that require measured anthropometrics or biomarkers. Use BRFSS and large survey-derived datasets for state-level aggregation and to ensure coverage of all U.S. states. When combining, harmonize variables (e.g., convert activity to minutes/week, compute BMI consistently), document all transformations, and use sample weights when required by survey design.
