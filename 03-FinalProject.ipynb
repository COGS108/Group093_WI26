## Abstract

This project investigates the association between self-reported weekly intentional exercise minutes and the prevalence of obesity, diabetes, and mental health outcomes among U.S. adults. While the benefits of physical activity are well-established at the individual level, we sought to examine whether these relationships hold at a broader, population-level scale and how they interact with demographic factors such as age and occupation. We analyzed two complementary datasets: a large-scale CDC BRFSS 2015 survey (n=253,680) providing population-level data on physical activity, BMI, diabetes status, and mental health days, and a smaller, individual-level dataset (n=300) with detailed exercise habits and self-reported mental health metrics. By categorizing exercise levels according to WHO guidelines (150+ minutes weekly), we explored relationships through visualizations including pairplots and boxplots.

Our analysis reveals a nuanced picture of the exercise-health relationship. In the BRFSS dataset, physically active adults demonstrated lower median BMIs and significantly fewer poor physical health days compared to inactive individuals. Diabetes and obesity rates increased predictably with age, with obesity rates rising from 22% in young adults to 38% by middle age, while diabetes rates climbed sharply after age 45, reaching 20% by ages 65-69. However, the mental health dataset painted a more complex picture. While higher exercise levels correlated with increased happiness scores, we found no consistent, strong association with reduced stress, anxiety, or depression. Notably, occupation emerged as a significant mediating factor: among employed individuals, higher exercise levels aligned with better mental health outcomes, whereas unemployed individuals in the high exercise category still reported elevated depression levels. This contradiction suggests that financial insecurity, loss of routine, and other socioeconomic factors may outweigh the mental health benefits of exercise. Overall, our findings indicate that while regular physical activity is generally associated with improved health outcomes, its benefits must be understood within a broader context of demographic and structural factors, and exercise alone cannot fully explain the variation in mental health across the U.S. adult population.

### Diabetes Risk Factors - BRFSS 2015

This dataset is a cleaned and consolidated version of the 2015 Behavioral Risk Factor Surveillance System (BRFSS) survey data, focusing on health indicators related to diabetes. The BRFSS is an annual, state-based, random-digit-dialed telephone survey conducted by the CDC, collecting data from over 400,000 Americans on their health-related risk behaviors, chronic health conditions, and use of preventative services. For this project, the dataset is valuable for exploring the relationship between physical activity (our key variable) and the prevalence of diabetes and obesity.

A. Important Metrics and Their Meaning

The dataset is available in three versions, but they all share the same core set of 21 feature variables. The key metrics relevant to our research question, along with their units and interpretation, are:

*   Diabetes Status (Target Variable): This is the primary outcome variable for diabetes.
    *   In `diabetes_012_health_indicators_BRFSS2015.csv`, this is a 3-class variable: `0` (no diabetes or only during pregnancy), `1` (prediabetes), and `2` (diabetes).
    *   In the two binary files, the variable `Diabetes_binary` is a 2-class variable where `0` indicates no diabetes and `1` indicates prediabetes or diabetes.
*   Physical Activity (`PhysActivity`): This is a binary variable directly relevant to our hypothesis. It indicates whether the respondent engaged in any exercise during the past 30 days, excluding their regular job. The units are binary: `0` (no) or `1` (yes). While it doesn't provide minutes per week, it serves as a high-level indicator of an active versus sedentary lifestyle.
*   Body Mass Index (`BMI`): This is a key metric for our obesity outcome. It is a calculated value based on the respondent's height and weight. The standard formula is `(weight in pounds / (height in inches)^2) * 703`. The unit is `kg/m²`, and it provides a continuous measure of body fat based on height and weight.
*   General Health (`GenHlth`): This is a self-reported, ordinal variable assessing the respondent's perception of their general health on a scale of 1 to 5, where:
    *   `1 = excellent`
    *   `2 = very good`
    *   `3 = good`
    *   `4 = fair`
    *   `5 = poor`
    This can serve as a proxy for overall well-being.
*   Mental Health (`MentHlth`): This variable directly measures the number of days the respondent experienced poor mental health, including stress, depression, and problems with emotions, during the past 30 days. It is reported on a scale from 1 to 30 days. This is a crucial variable for our mental health outcome.
*   Demographic & Socioeconomic Variables: The dataset includes other important variables for stratified analysis, such as:
    *   Age (`Age`): A 13-level categorical variable, not a continuous age. For example, `1 = 18-24`, `9 = 60-64`, `13 = 80 or older`.
    *   Sex (`Sex`): Binary, with `0 = female` and `1 = male`.
    *   Income (`Income`): An 8-level ordinal scale of annual household income (e.g., `1 = less than $10,000`, `8 = $75,000 or more`).
    *   Education (`Education`): A 6-level ordinal scale of education level (e.g., `1 = Never attended school or only kindergarten`, `6 = College 4 years or more`).

B. Major Concerns with the Dataset

While this is a large and well-structured dataset, there are several important limitations and concerns to consider for our analysis:

1.  Physical Activity is Not a Continuous Variable: The most significant limitation for our specific research question is that physical activity is recorded as a binary variable (`PhysActivity`). It only indicates whether any activity occurred in the past 30 days, not the frequency, duration, or intensity of that activity. This prevents us from calculating "average weekly intentional exercise (measured in minutes per week)" as stated in our research question and hypothesis. We can only compare those who did any activity versus those who did none.

2.  Data is from 2015: The survey data is nearly a decade old (from 2015). Health behaviors and population health trends can change over time, so the findings may not perfectly reflect current associations.

3.  Self-Reported Data and Associated Biases: All variables, including `BMI`, `MentHlth`, `PhysActivity`, and `GenHlth`, are based on self-report. This introduces potential biases:
    *   Recall Bias: Respondents may not accurately remember their exact number of poor mental health days or physical activity over the past 30 days.
    *   Social Desirability Bias: Respondents may underreport weight (affecting `BMI`) or overreport physical activity to present themselves in a better light.

4.  Class Imbalance: The primary dataset (`diabetes_012_health_indicators_BRFSS2015.csv`) has a significant class imbalance, meaning there are far more respondents without diabetes than with prediabetes or diabetes. This could bias a predictive model if not handled properly. The balanced 50/50 split version is provided to mitigate this for classification tasks.

5.  Aggregated and Categorical Variables: Some variables are provided in categorical ranges (e.g., `Age`, `Income`) rather than as continuous values. This results in a loss of granularity and statistical power. For instance, we cannot pinpoint an exact age but only a range.

This dataset is excellent for its large, population-based sample from the U.S. and its rich set of health and demographic indicators. However, for our project, we must critically address the binary nature of the `PhysActivity` variable, as it is a crude measure that does not align perfectly with our research question about exercise minutes.

### Exploratory Data Analysis

The following code loads the processed datasets from **data/02-processed**, computes `Weekly_Exercise_Minutes`, assigns `Exercise_Category` based on WHO guidelines, and creates standardized age groups to enable cross-dataset comparisons. The resulting **clean_combined dataset** is then used throughout the EDA below.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
clean_am_sr = pd.read_csv('data/02-processed/mental_health_physical_activity_clean.csv')
clean_diabetes = pd.read_csv('data/02-processed/diabetes_clean.csv')

# Calculate weekly exercise minutes
clean_am_sr['Weekly_Exercise_Minutes'] = (
    clean_am_sr['Exercise_Frequency'] * clean_am_sr['Exercise_Duration']
)

# Create exercise categories based on WHO guidelines
def categorize_exercise(minutes):
    """Categorizes weekly exercise minutes into groups based on WHO guidelines."""
    if minutes < 60:
        return 'Minimal (<60 min)'
    elif minutes < 150:
        return 'Insufficient (60-149 min)'
    elif minutes < 300:
        return 'Adequate (150-299 min)'
    else:
        return 'High (300+ min)'

clean_am_sr['Exercise_Category'] = clean_am_sr['Weekly_Exercise_Minutes'].apply(categorize_exercise)

# Create age groups that match BRFSS 5-year bins
def assign_age_group(age):
    """Converts numerical age into standardized 5-year age group bins based on BRFSS categories."""
    if age < 25: return '18-24'
    elif age < 30: return '25-29'
    elif age < 35: return '30-34'
    elif age < 40: return '35-39'
    elif age < 45: return '40-44'
    elif age < 50: return '45-49'
    elif age < 55: return '50-54'
    elif age < 60: return '55-59'
    elif age < 65: return '60-64'
    elif age < 70: return '65-69'
    elif age < 75: return '70-74'
    elif age < 80: return '75-79'
    else: return '80 or older'

clean_am_sr['Age_Group'] = clean_am_sr['Age'].apply(assign_age_group)

# Create composite mental health score (reverse-coded negative indicators)
# Higher score = better mental health
clean_am_sr['Mental_Health_Composite'] = (
    clean_am_sr['Happiness_Level'] -
    (clean_am_sr['Stress_Level'] + clean_am_sr['Anxiety_Level'] + clean_am_sr['Depression_Level']) / 3
)

# Create gender labels and BMI categories for obesity analysis
clean_diabetes['Gender'] = clean_diabetes['Sex'].map({0: 'Female', 1: 'Male'})
clean_diabetes['BMI_Category'] = pd.cut(
    clean_diabetes['BMI'],
    bins=[0, 18.5, 25, 30, np.inf],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese']
)
clean_diabetes['Obesity'] = (clean_diabetes['BMI'] >= 30).astype(int)

print("Mental-health dataset:", clean_am_sr.shape)
print("BRFSS dataset:        ", clean_diabetes.shape)


# Combine datasets by Age_Group × Gender

mh_agg = clean_am_sr.groupby(['Age_Group', 'Gender']).agg(
    MH_Avg_Exercise_Min   = ('Weekly_Exercise_Minutes', 'mean'),
    MH_Pct_WHO_Met        = ('Weekly_Exercise_Minutes', lambda x: (x >= 150).mean() * 100),
    MH_Avg_Happiness      = ('Happiness_Level', 'mean'),
    MH_Avg_Stress         = ('Stress_Level', 'mean'),
    MH_Avg_Anxiety        = ('Anxiety_Level', 'mean'),
    MH_Avg_Depression     = ('Depression_Level', 'mean'),
    MH_Avg_Composite      = ('Mental_Health_Composite', 'mean'),
    MH_N                  = ('Age', 'count')
).round(2).reset_index()

brfss_agg = clean_diabetes.groupby(['Age', 'Gender']).agg(
    BRFSS_Pct_Active      = ('PhysActivity', lambda x: x.mean() * 100),
    BRFSS_Avg_BMI         = ('BMI', 'mean'),
    BRFSS_Pct_Obese       = ('Obesity', lambda x: x.mean() * 100),
    BRFSS_Pct_Diabetes    = ('Diabetes_binary', lambda x: x.mean() * 100),
    BRFSS_Avg_Poor_MH_Days= ('MentHlth', 'mean'),
    BRFSS_Avg_GenHlth     = ('GenHlth', 'mean'),
    BRFSS_N               = ('Sex', 'count')
).round(2).reset_index().rename(columns={'Age': 'Age_Group'})


clean_combined = mh_agg.merge(brfss_agg, on=['Age_Group', 'Gender'], how='inner')
print(f"\nCombined dataset: {clean_combined.shape[0]} demographic groups, {clean_combined.shape[1]} columns")
clean_combined
```

## Discussion and Conclusion

After completing this project, we learned that higher intentional weekly exercise minutes are generally associated with better mental and physical health for adults in the United States. The strength of these associations varied depending on the variable we were measuring. In the Mental Health dataset, adults with higher weekly exercise minutes showed higher levels of happiness and slightly lower stress and depression levels, but there was still an overlap between the different exercise groups. This demonstrates that exercise alone does not fully explain mental health outcomes a finding that extends the existing literature by quantifying the limitations of exercise as a predictor of mental well-being.

The BRFSS Diabetes dataset revealed that more physically active adults tended to have slightly lower BMIs and fewer reported poor physical and mental health days compared to their less physically active counterparts. These findings align with prior research from the CDC documenting the protective effects of physical activity against chronic disease. Our demographic comparisons showed that diabetes and obesity rates increase predictably with age—obesity rates rising from 22% in young adults to 38% by middle age, while diabetes rates climb sharply after age 45 to reach 20% by ages 65-69. Mental health, on the other hand, varied more across age groups, with young adults reporting the highest number of poor mental health days despite also reporting the highest exercise participation rates. This apparent paradox suggests that while exercise benefits mental health, it cannot fully offset the unique stressors facing younger populations.

A notable contradiction emerged in our analysis of unemployed individuals. Contrary to our hypothesis and to prior research suggesting that higher physical activity consistently improves mental health we found that even unemployed individuals in the high exercise category (300+ minutes weekly) still reported elevated levels of depression. To investigate further, we compared this group with employed individuals in the same exercise category and found that employed high exercise individuals did report lower depression levels, aligning with our original hypothesis. This comparison strongly suggests that occupation plays a significant role in mental health outcomes. For unemployed individuals, factors such as financial insecurity, loss of routine, and social isolation may outweigh the mental health benefits that physical activity can provide. This contradiction highlights an important limitation of our research: our cross-sectional data captures only a single point in time and cannot account for the duration or consistency of physical activity over a person's lifetime.

Several limitations warrant consideration when interpreting our findings. First, both datasets rely on self-reported measures, introducing potential recall and social desirability bias—participants may overreport exercise and underreport weight or poor health days. Second, the mental health dataset's small sample size (n=300) limits generalizability to the broader U.S. adult population, particularly when examining demographic subgroups. Third, our analysis is cross-sectional, allowing us to identify associations but not causation; the relationships we observe could reflect reverse causality (e.g., healthier individuals being more able to exercise) or unmeasured confounding variables such as diet quality, sleep patterns, and healthcare access. Fourth, the BRFSS data is from 2015, nearly a decade old, and may not reflect current trends, especially given the significant lifestyle changes following the COVID-19 pandemic. Fifth, the binary nature of physical activity measurement in the BRFSS dataset prevents fine-grained analysis of exercise intensity and duration, potentially masking important dose-response relationships.

What have we learned from bringing these results together? Physical activity is undeniably beneficial for health, but it is not a panacea. Our findings suggest that while exercise is associated with improved physical health markers across all age groups, its relationship with mental health is more complex and highly context-dependent. The mediating role of occupation in mental health outcomes underscores the importance of considering structural and socioeconomic factors when designing public health interventions. Simply encouraging more exercise may not be sufficient for populations facing unemployment, financial stress, or social isolation.

If this work were to continue beyond this class, several next steps would be valuable. First, incorporating longitudinal data would allow us to track individuals over time and better understand causal relationships—for example, does exercise improve mental health, or do mentally healthier individuals maintain exercise habits? Second, collecting more granular data on exercise intensity, type (aerobic vs. strength training), and consistency would enable dose-response analysis to identify optimal exercise prescriptions for different health outcomes. Third, integrating additional demographic variables such as income, education, and healthcare access would help disentangle the complex interplay between socioeconomic factors and health behaviors. Fourth, conducting state-level spatial analysis could identify geographic disparities in both exercise access and health outcomes, informing targeted public health interventions. Finally, a predictive model—such as logistic regression classifying diabetes, obesity, or depression risk based on exercise minutes and demographic factors—could help identify at-risk populations for early intervention.
